Ariketa 1.

### A. Aurkezpena: finka ditzagun ideiak.

>Sistema hidrostatikoa aztertuko duzu.  
 Horietan errezenatakoa: gasa (ez dut esan gas ideala denik...)  

>Gasa sartu suzu pistoi diatermikoa itxi duen zilindro horizontal batean.  
Zilindroa dago bero-iturri batekin ukipen termikoan.  
Berebat, pistoiaren gainean presio konstantea eragiten da.  

    1. Egin irudia. Detaile guztiek balioa dute.  

       Demagun era infinitesimalean aldatzen dugula bolumena: $dV$ eragiten dugu.  
       Egin dezagun hauxe: 
## $$dV < 0$$  
       Nolakoa izango da sisteman gertatuko den $dp$ aldaketa deiferentziala?  
       
       Eta sistemak nola erreakzionatuko du?  
       Zelan aldatuko du sistemak presioa?
       
## $$d p $$##       
       
       Azken horri ***sistemaren erantzuna*** deritzo.

### C. Le'Chatelier-en eta Braun-en Printzipioaren edukin fisikoa:  

    3. Aztertu zer eragin daukan bolumenaren aldaketa diferentzialak,  
       sistemaren tenperaturan  
       
       Hots, oraingoan ez duzu aztertu behar aldatu dugun  
       aldagai estentsiboarekin lotutako aldagai intentsiboaren aldaketa
       baizik eta beste aldagai intentsiaboan eragin den aldaketa.  
       
       Horren arabera, sistemak xurgatuko ala kanporatuko al du beroa?  
       
       Beraz, zer egingo du sistemak?  
       Zein izango da sistemaren erantzuna?  
       
       Lortu honako hau:
       
## $$ dp= dp\left(C_{V}, \kappa_{T}, \alpha, \delta Q\right)$$##
